```python
import pycistem
import matplotlib.pyplot as plt
import numpy as np
image = pycistem.Image()
```

```python
image.QuickAndDirtyReadSlice("/scratch/bern/elferich/denoising/denoise_model2/im63_DW.mrc",1)
```

```python
plt.imshow(np.array(image,copy=False),cmap="Greys_r",interpolation="bicubic")
```

```python
smallest_dimension = np.min(np.array(image,copy=False).shape)
square_image = pycistem.Image()
square_image.Allocate(smallest_dimension,smallest_dimension,True)
image.ClipInto(square_image,0.0,False,0.0,0,0,0)
```

```python
plt.imshow(np.array(square_image,copy=False),cmap="Greys_r",interpolation="bicubic")
```

```python
square_image.ForwardFFT(True)
square_image.ZeroCentralPixel()
power_spectrum = pycistem.Image()
power_spectrum.Allocate(smallest_dimension,smallest_dimension,True)
square_image.ComputeAmplitudeSpectrumFull2D(power_spectrum,False,1.0)
```

```python
plt.imshow(np.array(power_spectrum,copy=False),cmap="Greys_r",interpolation="bicubic",vmax=0.00003)
```
